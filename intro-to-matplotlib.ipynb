### making two or more in same plot

```python
#subploting ->1

fig,((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows=2,
                                        ncols=2,
                                        figsize=(10,10))

ax1.plot(x,x/2);
ax2.scatter(np.random.random(10),np.random.random(10))
ax3.bar(list(butter_prices.keys()),list(butter_prices.values()));
ax4.hist(np.random.randn(100000));
```

```python
#2nd option
fig,ax = plt.subplots(nrows=2,
                      ncols=2,
                      figsize=(10,10))

ax[0,0].plot(x,x/2);
ax[0,1].scatter(np.random.random(10),np.random.random(10))
ax[1,0].bar(list(butter_prices.keys()),list(butter_prices.values()));
ax[1,1].hist(np.random.randn(100000));
```

## Heart disease Excercise

```python
df = pd.read_csv("/kaggle/input/heart-disease-dataset/heart.csv")
df.head()
```

```python
df_50 = df[df["age"]>50]
df_50.head()
```

```python
fig,ax = plt.subplots(figsize=(10,6))
scatter = ax.scatter(x=df_50["age"],
                     y = df_50["chol"],
                    c=df_50["target"])

ax.set(title="Heart Disease and Cholesterol Levels",
           xlabel="Age",
           ylabel = "Cholesterol");

ax.legend(*scatter.legend_elements(), title="Target")
plt.axhline(y=df_50["chol"].mean(),linestyle="--",label="Average",color="red");
```

```python
# Create the plot
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the data
scatter = ax.scatter(df_50["age"], 
                     df_50["chol"], 
                     c=df_50["target"],
                     cmap="winter") # changed cmap parameter

# Customize the plot
ax.set(title="Heart Disease and Cholesterol Levels",
       xlabel="Age",
       ylabel="Cholesterol");
ax.legend(*scatter.legend_elements(), title="Target")

# Add a meanline
ax.axhline(df_50["chol"].mean(),
           linestyle="--",
           color="red"); # changed color parameter
```
